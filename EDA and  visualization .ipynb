#  Visualizing Basic Analysis of the data using plots

```python
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Plotting a Barplot 

```python
plt.figure(figsize=(10,6))
sns.barplot(x=laptops_df['Model_Price'],y=laptops_df['Product_Rating'])
plt.title("Price vs Rating")
```

## We can conclude from here that products with higher prices have higher ratings to some extent.

```python
plt.figure(figsize=(10,6))
sns.barplot(x=laptops_df['Product_Rating'],y=laptops_df['Model_Price'])
plt.title("Rating vs Price")
```

## Plotting a Histrogram

```python
laptops_df['Model_Price'].hist(figsize=(8,6),bins=15)
```

## we can see the range of  purchasing laptops from  40,000 to 60,0000 INR is more 

```python
laptops_df['Product_Rating'].hist(figsize=(8,6),bins=15)
```

## we can see the rating range  from 4 to 4.8  is more 

## Plotting a Boxplot

```python
sns.boxplot(x=laptops_df['Model_Price'])
plt.title("Box plot")
```

## As we can see there is one outlier where the price range is very high

```python
sns.boxplot(x=laptops_df['Product_Rating'])
plt.title("Box plot")
```

## As we can see there are two outlier where the rating range is very low 

## Plotting a lineplot

```python
sns.lineplot(x=laptops_df['Model_Price'],y=laptops_df['Product_Rating'])
plt.title("Model_Price vs Product_Rating")
```

## As price increasing rating  also increasing along with price 

```python
sns.lineplot(x=laptops_df['Product_Rating'],y=laptops_df['Model_Price'])
plt.title("Product_Rating vs Model_Price")
```

## Plotting Densityplot

```python
columns=['Model_Price','Product_Rating']
for i in columns:
    sns.kdeplot(laptops_df[i],shade=True)
    plt.show()
```

 ## Plotting Histograms with Densityplot

```python
columns=['Model_Price','Product_Rating']
for i in columns:
    sns.distplot(laptops_df[i],bins=20)
    plt.show()
```

## Plotting Scatter plot

```python
plt.scatter(laptops_df['Model_Price'],laptops_df['Product_Rating'])
plt.title('PRICE vs RATING')
```
